### 변수의 필요성 

데이터 (234342, 34422, 4566 등)를 재사용할 필요가 있다.  
이것이 변수가 필요한 이유다.  
변수는 데이터를 담는 일종의 데이터통이라고 볼 수 있다.  
  
물이 있다고 하자. 그럼 물을 담을 물통이 필요하다.  
물 = 데이터, 물통 = 변수 라고 보면 된다.  
  
그럼 이제 친구에게 물 좀 달라고 부탁을 했다고 생각해보자.  
친구가 손에 물을 떠오길 기대하는가, 아니면 물이 담긴 물통을 가져와주길 기대하는가.  
  
**우리는 이미 데이터를 다룰 때, 데이터를 변수에 담아 변수를 다루는 것에 익숙하다.** 
  
이제 남은 일은 물통에 이름을 붙이는 일이다.  
물통마다 이름이 다다르고, 담긴 물도 다 다르다고 생각해보자.  
삼다수라는 이름의 물통에는 삼다수가, 옥수수수염차라는 이름의 물통에는 옥수수수염차가 있다.  
그리고 우리는 삼다수가 먹고싶다.  
그럼 친구에게 부탁을 할 때, 삼다수를 가져다 달라고 부탁을 할 것이다.  
  
이렇게, **우리는 변수를 다룰 때, 변수의 이름을 정하고 변수의 이름으로 데이터를 다루게 된다.**  
여기서 중요한 것은 **변수의 이름에 담긴 데이터의 정보가 직관적으로 잘 표현되어야 좋다는 사실이다.**

### 변수이름 작성법
- 직관성만 기억하면 끝
    - 문자 또는 언더바(_)로 시작 (숫자로 시작 x)
    - 언더바로 시작하는 변수명도 되도록 사용하지 않는 것이 좋음

    - 대소문자 구분 (하지만, 이를 이용해서 변수를 구별하는 것은 좋지 않음)
    - 특수문자, 공백, 한글 포함 금지
