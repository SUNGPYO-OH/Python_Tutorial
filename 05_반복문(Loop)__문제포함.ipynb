## for

    기본 : 'FOR문은 리스트를 반복하는 것이다'

for (원소) in (리스트):   
->->->->(원하는 코드)

```python
foods = ['볶음밥', '비빔밥', '돈가스']

for food in foods: # for 원소를 대입받을 변수명 in 리스트:
    print(food)    # 반복할 실행 코드
```

### 반복 시행 시 동시에 여러개의 리스트를 함께 활용하고 싶다면

출력 결과

볶음밥 먹고, 아메리카노  
비빔밥 먹고, 딸기라떼  
돈가스 먹고, 버블티  

```python
foods = ['볶음밥', '비빔밥', '돈가스']
drinks = ['아메리카노', '딸기라떼', '버블티']

for food, drink in zip(foods, drinks):
    print(food, '먹고,', drink)
```

```python
foods = ['볶음밥', '비빔밥']
drinks = ['아메리카노', '딸기라떼', '버블티']

for food, drink in zip(foods, drinks):
    print(food, '먹고,', drink)
```

```python
from itertools import zip_longest

foods = ['볶음밥', '비빔밥']
drinks = ['아메리카노', '딸기라떼', '버블티']

for food, drink in zip_longest(foods, drinks):
    print(food, '먹고,', drink)
```


#### for i in range(시작점, 끝점+1, 변화간격) 용법

- **i 는 변수다. (i am variables)**
- 시작점은 안쓰면 0, 변화간격은 안쓰면 +1
- 끝점+1은 반드시 적어줘야함. 

```python
for i in range(10): #시작점, 변화간격 안씀 -> 0부터 9까지 +1씩 변화
    print(i)
```

```python
# 반복문을 이용해서 리스트 만들기 
num_list = []
for number in range(10):
    num_list.append(number)
print(num_list)

# 사실상 위와 같은 리스트생성은 반복문을 사용할 필요가 없다.
# 반복문 사용하지 않고 위와 같은 리스트 만들기
num_list = list(range(10))
print(num_list)
```

#### continue 용법
해당 루프만 중단하고 싶을 때,  
바로 다음루프로 진행시키고 싶을 때


```python
for i in range(10):
    if i%2 == 0: # 0이거나 짝수일 때
        continue
    print(i)
```

```python
### 0부터 99까지 반복시행하면서 
### 3 의 배수만 출력하도록 
### continue를 이용해 코드를 작성해보자.

```

```python
#@title
for i in range(100):    
    if i%3 != 0 or i == 0:
        continue
    print(i)
```

#### 중첩 for문 용법

```python
# 101호, 102호, ... 105호 
# ...
# 1001호 ... 1005호까지

### 5번씩 10회 반복
for layer in range(10): # 층
    for ho in range(5): # 호
        print(layer,'층', ho,'호') # 5바퀴 씩 10바퀴 = 총 50회 수행
    print('\n')
```

3중 중첩 이상은 최대한 회피한다.  
시간 복잡도가 너무 높아진다.

```python
for i in range(100):
    for j in range(100):
        for k in range(100):
            print(i, j, k) # 100 * 100 * 100 = 100만회 실행
```

## While

#### while(조건) : 조건이 참인 동안 loop

##### 특정 조건이 참인 동안 계속 반복하는 코드

```python
cnt = 1

while cnt < 10:
    print(cnt)
    cnt += 1 # cnt = cnt + 1
```

##### 무한반복하는 코드

```python
while 1: # while 조건 : 조건의 return True 라면 종속절 실행 -> loop
    print('hello')
```

##### 무한반복문과 break를 이용

```python
#무한반복
count = 0
while 1:
    count += 1 #-> count = count + 1
    if count >= 100:
        print('loop finished')
        break #바로 상위의 loop 1개 종료
    print(count)
```

### 이 함수는 어떤 함수일까요?

```python
a = 0

while a < 10:
    a += 1
    if a % 2 == 0:
        continue
    print(a)
```

    - 짝수이면 continue

- continue는 while문의 맨 처음(조건문 a <10)으로 돌아가게 합니다!
